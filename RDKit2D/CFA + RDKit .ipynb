```python
def pred_prob_to_score(pred_prob):
    res = []
    for i in range(len(pred_prob)):
        res.append(pred_prob[i][1])
    res = np.array(res)
    return res
```

```python
group = admet_group(path = 'data/')
mae_datasets = ['Caco2_Wang','Lipophilicity_AstraZeneca','Solubility_AqSolDB','PPBR_AZ','LD50_Zhu']
spearman_datasets = ['VDss_Lombardo','Half_Life_Obach','Clearance_Hepatocyte_AZ','Clearance_Microsome_AZ']
auroc_datasets = ['Bioavailability_Ma','HIA_Hou','Pgp_Broccatelli','BBB_Martins','CYP3A4_Substrate_CarbonMangels','hERG','AMES','DILI']
auprc_datasets = ['CYP2C9_Veith','CYP2D6_Veith','CYP3A4_Veith','CYP2C9_Substrate_CarbonMangels','CYP2D6_Substrate_CarbonMangels']
names = ['HIA_Hou', 'Pgp_Broccatelli', 'Bioavailability_Ma', 'BBB_Martins', 'Clearance_Hepatocyte_AZ', 'Clearance_Microsome_AZ', 'hERG', 'DILI', 'CYP3A4_Substrate_CarbonMangels', 'CYP2C9_Substrate_CarbonMangels', 'CYP2D6_Substrate_CarbonMangels']
models = ['xgb','rf','svm','adb','cnn']

for name in names:
    pred_valid_list_xgb = []
    pred_valid_list_rf = []
    pred_valid_list_svm = []
    pred_valid_list_adb = []
    pred_valid_list_cnn = []
            
    pred_test_list_xgb = []
    pred_test_list_rf = []
    pred_test_list_svm = []
    pred_test_list_adb = []
    pred_test_list_cnn = []

    best_params_list_xgb = []
    best_params_list_svm = []
    
    pred_valid_xgb = pd.DataFrame()
    pred_valid_rf = pd.DataFrame()
    pred_valid_svm = pd.DataFrame()
    pred_valid_adb = pd.DataFrame()
    pred_valid_cnn = pd.DataFrame()

    pred_test_xgb = pd.DataFrame()
    pred_test_rf = pd.DataFrame()
    pred_test_svm = pd.DataFrame()
    pred_test_adb = pd.DataFrame()
    pred_test_cnn = pd.DataFrame()

    benchmark = group.get(name)
    train_val, test = benchmark['train_val'], benchmark['test']

    for seed in [1, 2, 3, 4, 5]:
        train, valid = group.get_train_valid_split(benchmark = name, split_type = 'default', seed = seed)
        nBits=200
        # train
        rdkit_des = [get_rdkit(smile) for smile in train['Drug']]
        rdkit_name = [f'Bit_{i}' for i in range(nBits)]
        rdkit_bits = [list(l) for l in rdkit_des]
        Y = train['Y']
        train = pd.DataFrame(rdkit_bits, index = train.Drug, columns=rdkit_name).reset_index(drop = False)
        train['Y'] = Y
        if train.isnull().sum().sum() >= 1:
            train = train.dropna(axis=1)
        # valid
        rdkit_des = [get_rdkit(smile) for smile in valid['Drug']]
        rdkit_name = [f'Bit_{i}' for i in range(nBits)]
        rdkit_bits = [list(l) for l in rdkit_des]
        Y = valid['Y']
        valid = pd.DataFrame(rdkit_bits, index = valid.Drug, columns=rdkit_name).reset_index(drop = False)
        valid['Y'] = Y
        if valid.isnull().sum().sum() >= 1:
            valid = valid.dropna(axis=1)
        # test
        rdkit_des = [get_rdkit(smile) for smile in test['Drug']]
        rdkit_bits = [list(l) for l in rdkit_des]
        Y = test['Y']
        test = pd.DataFrame(rdkit_bits, index = test.Drug, columns=rdkit_name).reset_index(drop = False)
        test['Y'] = Y
        if test.isnull().sum().sum() >= 1:
            test = test.drop(axis=1)

        train_X = train.drop(columns = ["Drug","Y"])
        train_y = train.Y
        valid_X = valid.drop(columns = ["Drug","Y"])
        globals()['valid_y_%s'%seed] = valid.Y
        test_X = test.drop(columns = ["Drug","Y"])
        test_y = test.Y

        if name in mae_datasets or spearman_datasets:
            #XGBoost + MCFP
            xgb_parameters = {'objective':['reg:squarederror'],
                    'booster':['gbtree','gblinear'],
                    'learning_rate': [0.1],
                    'max_depth': [7,10,15,20],
                    'min_child_weight': [10,15,20,25],
                    'colsample_bytree': [0.8, 0.9, 1],
                    'n_estimators': [300,400,500,600],
                    "reg_alpha"   : [0.5,0.2,1],
                    "reg_lambda"  : [2,3,5],
                    "gamma"       : [1,2,3]}

            xgb_model = XGBRegressor()
            grid_obj_xgb = RandomizedSearchCV(xgb_model, xgb_parameters, cv=5, n_iter=15, scoring='neg_mean_absolute_error', verbose=5, n_jobs=1)
            grid_obj_xgb.fit(train_X, train_y, verbose = 1)
            y_pred_valid_xgb = grid_obj_xgb.predict(valid_X)
            y_pred_test_xgb = grid_obj_xgb.predict(test_X)
            pred_valid_xgb = pd.concat([pred_valid_xgb, pd.Series(y_pred_valid_xgb)], ignore_index=True, axis=1)
            pred_valid_list_xgb.append(pred_valid_xgb)
            pred_test_xgb['seed_%s' % seed] = y_pred_test_xgb
            pred_test_list_xgb.append(pred_test_xgb)
            bp_xgb = grid_obj_xgb.best_params_
            best_params_list_xgb.append(bp_xgb)

            #Random Forest + MCFP
            rf_model = RandomForestRegressor()
            rf_model.fit(train_X, train_y)
            y_pred_valid_rf = rf_model.predict(valid_X)
            y_pred_test_rf = rf_model.predict(test_X)
            pred_valid_rf = pd.concat([pred_valid_rf, pd.Series(y_pred_valid_rf)], ignore_index=True, axis=1)
            pred_valid_list_rf.append(pred_valid_rf)
            pred_test_rf['seed_%s' % seed] = y_pred_test_rf
            pred_test_list_rf.append(pred_test_rf)

            #SVM + MCFP
            svm_parameters = {
                'C': [0.1, 1, 10, 100, 1000],
                'epsilon': [0.0001, 0.001, 0.01, 0.1, 0.5],
                'gamma': [0.0001, 0.001, 0.01, 0.1, 1]
            }
            svm_model = SVR(kernel="rbf")
            grid_obj_svm = RandomizedSearchCV(svm_model, svm_parameters, cv=5, n_iter=15, scoring='neg_mean_absolute_error', verbose=5, n_jobs=1)
            grid_obj_svm.fit(train_X, train_y)
            y_pred_valid_svm = grid_obj_svm.predict(valid_X)
            y_pred_test_svm = grid_obj_svm.predict(test_X)
            pred_valid_svm = pd.concat([pred_valid_svm, pd.Series(y_pred_valid_svm)], ignore_index=True, axis=1)
            pred_valid_list_svm.append(pred_valid_svm)
            pred_test_svm['seed_%s' % seed] = y_pred_test_svm
            pred_test_list_svm.append(pred_test_svm)
            bp_svm = grid_obj_svm.best_params_
            best_params_list_svm.append(bp_svm)
            
            #AdaBoost + MCFP
            DTR = DecisionTreeRegressor(max_depth=4)
            adb_model = AdaBoostRegressor(n_estimators=300, base_estimator=DTR, learning_rate=1)
            adb_model.fit(train_X, train_y)
            y_pred_valid_adb = adb_model.predict(valid_X)
            y_pred_test_adb = adb_model.predict(test_X)
            pred_valid_adb = pd.concat([pred_valid_adb, pd.Series(y_pred_valid_adb)], ignore_index=True, axis=1)
            pred_valid_list_adb.append(pred_valid_adb)
            pred_test_adb['seed_%s' % seed] = y_pred_test_adb
            pred_test_list_adb.append(pred_test_adb)

            #CNN + MCFP
            train_X_cnn = train_X.to_numpy()
            valid_X_cnn = valid_X.to_numpy()
            test_X_cnn = test_X.to_numpy()
            train_X_cnn = train_X_cnn.reshape(train_X_cnn.shape[0], train_X_cnn.shape[1], 1)
            valid_X_cnn = valid_X_cnn.reshape(valid_X_cnn.shape[0], valid_X_cnn.shape[1], 1)
            test_X_cnn = test_X_cnn.reshape(test_X_cnn.shape[0], test_X_cnn.shape[1], 1)
            cnn_model = Sequential()
            cnn_model.add(Conv1D(32, 2, activation="relu", input_shape=(1024,1)))
            cnn_model.add(Flatten())
            cnn_model.add(Dense(64, activation="relu"))
            cnn_model.add(Dense(1))
            cnn_model.compile(loss="mse", optimizer="adam")
            cnn_model.fit(train_X_cnn, train_y, batch_size=12, epochs=10, verbose=0)
            y_pred_valid_cnn = cnn_model.predict(valid_X_cnn)
            y_pred_test_cnn = cnn_model.predict(test_X_cnn)
            y_pred_valid_cnn_copy = []
            for i in range(len(y_pred_valid_cnn)):
                y_pred_valid_cnn_copy.append(y_pred_valid_cnn[i][0])
            y_pred_test_cnn_copy = []
            for i in range(len(y_pred_test_cnn)):
                y_pred_test_cnn_copy.append(y_pred_test_cnn[i][0])
            pred_valid_cnn = pd.concat([pred_valid_cnn, pd.Series(np.array(y_pred_valid_cnn_copy))], ignore_index=True, axis=1)
            pred_valid_list_cnn.append(pred_valid_cnn)
            pred_test_cnn['seed_%s' % seed] = np.array(y_pred_test_cnn_copy)
            pred_test_list_cnn.append(pred_test_cnn)
        else:
            #XGBoost + MCFP
            xgb_parameters = {'objective':['binary:logistic'],
            'booster':['gbtree','gblinear'],
            'learning_rate': [0.1],
            'max_depth': [7,10,15,20],
            'min_child_weight': [10,15,20,25],
            'colsample_bytree': [0.8, 0.9, 1],
            'n_estimators': [300,400,500,600],
            "reg_alpha"   : [0.5,0.2,1],
            "reg_lambda"  : [2,3,5],
            "gamma"       : [1,2,3]}

            xgb_model = XGBClassifier()
            grid_obj_xgb = RandomizedSearchCV(xgb_model, xgb_parameters, cv=5, n_iter=15, scoring = 'roc_auc', verbose=5, n_jobs=1)
            grid_obj_xgb.fit(train_X, train_y, verbose = 1)
            y_pred_valid_xgb = grid_obj_xgb.predict_proba(valid_X)
            y_pred_test_xgb = grid_obj_xgb.predict_proba(test_X)
            pred_valid_xgb = pd.concat([pred_valid_xgb,pd.Series(pred_prob_to_score(y_pred_valid_xgb))], ignore_index=True, axis=1)
            pred_valid_list_xgb.append(pred_valid_xgb)
            pred_test_xgb['seed_%s' % seed] = pred_prob_to_score(y_pred_test_xgb)
            pred_test_list_xgb.append(pred_test_xgb)
            bp_xgb = grid_obj_xgb.best_params_
            best_params_list_xgb.append(bp_xgb)

            #Random Forest + MCFP
            rf_model = RandomForestClassifier()
            rf_model.fit(train_X, train_y)
            y_pred_valid_rf = rf_model.predict_proba(valid_X)
            y_pred_test_rf = rf_model.predict_proba(test_X)
            pred_valid_rf = pd.concat([pred_valid_rf,pd.Series(pred_prob_to_score(y_pred_valid_rf))], ignore_index=True, axis=1)
            pred_valid_list_rf.append(pred_valid_rf)
            pred_test_rf['seed_%s' % seed] = pred_prob_to_score(y_pred_test_rf)
            pred_test_list_rf.append(pred_test_rf)

            #SVM + MCFP
            svm_parameters = {
                'C': [0.1, 1, 10, 100, 1000],
                'gamma': [0.0001, 0.001, 0.01, 0.1, 1]
            }
            svm_model = SVC(kernel="rbf", probability=True)
            grid_obj_svm = RandomizedSearchCV(svm_model, svm_parameters, cv=5, n_iter=15, verbose=5, n_jobs=1)
            grid_obj_svm.fit(train_X, train_y)
            y_pred_valid_svm = grid_obj_svm.predict_proba(valid_X)
            y_pred_test_svm = grid_obj_svm.predict_proba(test_X)
            pred_valid_svm = pd.concat([pred_valid_svm,pd.Series(pred_prob_to_score(y_pred_valid_svm))], ignore_index=True, axis=1)
            pred_valid_list_svm.append(pred_valid_svm)
            pred_test_svm['seed_%s' % seed] = pred_prob_to_score(y_pred_test_svm)
            pred_test_list_svm.append(pred_test_svm)
            bp_svm = grid_obj_svm.best_params_
            best_params_list_svm.append(bp_svm)

            # AdaBoost + MCFP
            DTC = DecisionTreeClassifier(max_depth=4)
            adb_model = AdaBoostClassifier(n_estimators=300, base_estimator=DTC, learning_rate=1)
            adb_model.fit(train_X, train_y)
            y_pred_valid_adb = adb_model.predict_proba(valid_X)
            y_pred_test_adb = adb_model.predict_proba(test_X)
            pred_valid_adb = pd.concat([pred_valid_adb,pd.Series(pred_prob_to_score(y_pred_valid_adb))], ignore_index=True, axis=1)
            pred_valid_list_adb.append(pred_valid_adb)
            pred_test_adb['seed_%s' % seed] = pred_prob_to_score(y_pred_test_adb)
            pred_test_list_adb.append(pred_test_adb)

            #CNN + MCFP
            train_X_cnn = train_X.to_numpy()
            valid_X_cnn = valid_X.to_numpy()
            test_X_cnn = test_X.to_numpy()
            train_X_cnn = train_X_cnn.reshape(train_X_cnn.shape[0], train_X_cnn.shape[1], 1)
            valid_X_cnn = valid_X_cnn.reshape(valid_X_cnn.shape[0], valid_X_cnn.shape[1], 1)
            test_X_cnn = test_X_cnn.reshape(test_X_cnn.shape[0], test_X_cnn.shape[1], 1)
            train_X = train_X.astype('float32')
            valid_X = valid_X.astype('float32')
            test_X = test_X.astype('float32')
            cnn_model = Sequential()
            cnn_model.add(Conv1D(32, 2, padding="valid", activation="relu", input_shape=(1024,1)))
            cnn_model.add(Flatten())
            cnn_model.add(Dense(64, activation="relu"))
            cnn_model.add(Dense(1,activation='sigmoid'))
            cnn_model.compile(loss=tf.keras.losses.BinaryCrossentropy(), optimizer="adam", metrics=[tf.keras.metrics.BinaryAccuracy(), tf.keras.metrics.FalseNegatives()])
            cnn_model.fit(train_X_cnn, train_y, batch_size=12, epochs=10, verbose=0)
            y_pred_valid_cnn = cnn_model.predict(valid_X_cnn)
            y_pred_test_cnn = cnn_model.predict(test_X_cnn)
            y_pred_valid_cnn_copy = []
            for i in range(len(y_pred_valid_cnn)):
                y_pred_valid_cnn_copy.append(y_pred_valid_cnn[i][0])
            y_pred_test_cnn_copy = []
            for i in range(len(y_pred_test_cnn)):
                y_pred_test_cnn_copy.append(y_pred_test_cnn[i][0])
            pred_valid_cnn = pd.concat([pred_valid_cnn,pd.Series(np.array(y_pred_valid_cnn_copy))], ignore_index=True, axis=1)
            pred_valid_list_cnn.append(pred_valid_cnn)
            pred_test_cnn['seed_%s' % seed] = np.array(y_pred_test_cnn_copy)
            pred_test_list_cnn.append(pred_test_cnn)

    dump(best_params_list_xgb, '%s_best_params_xgb.joblib' % name)
    dump(best_params_list_svm, '%s_best_params_svm.joblib' % name)

    for model in models:
        globals()['pred_valid_%s' % model].rename(columns = {0:'seed_1',1:'seed_2',2:'seed_3',3:'seed_4',4:'seed_5'})
        globals()['pred_valid_%s' % model].to_csv('%s_predictions_val_%s.csv' % (name,model), index = False)
        globals()['pred_test_%s' % model].to_csv('%s_predictions_test_%s.csv' % (name,model), index = False)
```
